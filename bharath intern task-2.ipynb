```python
data.replace({'Sex':{"male":0,"female":1},"Embarked":{"Q":0,"S":1,"C":2}},inplace=True)
data.head()
```

```python
x=data.drop(["PassengerId","Name","Survived","Ticket"],axis=1)
y=data["Survived"]
```

```python
x
```

```python
y
```

```python
#Split the data from training and testing
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer instance
imputer = SimpleImputer(strategy='mean')

# Fit and transform on x_train to replace missing values
x_train_imputed = imputer.fit_transform(x_train)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a scaler instance
scaler = StandardScaler()

# Fit and transform on x_train to scale features
x_train_scaled = scaler.fit_transform(x_train_imputed)
```

```python
import numpy as np

print(np.isnan(x_train_scaled).sum())
print(np.isinf(x_train_scaled).sum())
print(np.isnan(y_train).sum())
print(np.isinf(y_train).sum())

```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(max_iter=1000)

model.fit(x_train_scaled,y_train)
```

```python
#Checking the accuracy
from sklearn.metrics import *
x_train_prediction=model.predict(x_train_scaled)
train_data_accuracy=accuracy_score(y_train,x_train_prediction)
print("Traning accuracy is:",train_data_accuracy)
```

```python
#Checking accuracy of testing
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(y_test,x_test_prediction)
print("Testing acuuracy is:",test_data_accuracy)
```
