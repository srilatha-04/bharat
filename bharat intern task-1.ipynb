```python
#DATA PREPROCESSING
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
df4=scaler.fit_transform(np.array(df2).reshape(-1,1))
df4
```

```python
#splitting the dataset into train and test spilt
train_size=int(len(df2)*0.75)
test_size=len(df2)-train_size
train_size,test_size
```

```python
trained_data=df2[1:train_size  :]
test_data=df2[train_size:len(df2)  :1]
```

```python
type(trained_data)
```

```python
type(test_data)
```

```python
trained_data
```

```python

import numpy as np
def  create_dataset(dataset,time_step=1):
    dataX,dataY=[],[]
    for i in range(len(dataset)-time_step-1):
        a=dataset[i:(i+time_step)  :1 ]
        dataX.append(a)
        dataY.append(dataset[i+time_step  :1])
    return np.array(dataX),np.array(dataY)
#reshape into x=t,t+1,t+2,t+3 and Y=t+4
time_step=100
x_train,y_train=create_dataset(trained_data,time_step)
x_test,y_test=create_dataset(test_data,time_step)
```

```python
print(x_train.shape),print(y_train.shape)
```

```python
print(x_test.shape),print(y_test.shape)
```

```python

x_train=x_train.reshape(x_train.shape[0],x_train.shape[1],1)
x_test=x_test.reshape(x_test.shape[0],x_test.shape[1],1)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
```

```python
!pip install tensorflow==2.5
```

```python
from tensorflow.keras.models import *
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(60,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(60,return_sequences=True))
model.add(LSTM(60))
model.add(Dense(1))
model.compile(loss='mean_squared_error'
              ,optimizer='adam')
model.summary()
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=6, batch_size=50, verbose=1)
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
#lets check the preduction and check performace metrix
train_predict=model.predict(x_train)
test_predict=model.predict(x_test)
print(train_predict),print(test_predict)
```

```python
#transform to original form
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
# Convert df2 to a numpy array and reshape for inverse_transform
original_data = scaler.inverse_transform(df2.to_numpy().reshape(-1, 1))

# Shifting train predictions for plotting
look_back=100
trainPlot = np.empty_like(df2)
trainPlot[: :] = np.nan
trainPlot[look_back : len(train_predict) + look_back  : ] = train_predict[:1]

# Shifting test predictions for plotting
testPlot = np.empty_like(df2)
testPlot[: :] = np.nan
testPlot[len(train_predict) + (look_back * 2) + 1 : len(df2) - 1 :] = test_predict[:1]

# Plotting Base line and predictions
plt.plot(original_data)
plt.plot(trainPlot)
plt.plot(testPlot)
plt.show()
```
